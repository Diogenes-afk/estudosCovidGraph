### importando biblioteca warning para n se preocupar com avisos de biblioteca 

```python
data['actives'] = data['Confirmed'] - data['Deaths'] - data['Recovered']
```

### criando coluna dentro do dataSet

```python
data['Country/Region'] = data['Country/Region'].replace('Mainland China', 'China')
```

```python
data[['Province/State']] = data[['Province/State']].fillna('')
data[['Confirmed', 'Deaths', 'Recovered', 'actives']] = data[['Confirmed', 'Deaths', 'Recovered', 'actives']].fillna(0)
```

### Limpeza das bases, o que for NaN troca para vazio e o que for vazio em colunas de float troca para 0.

#### Foi alterado o nome do pais China na linha 13

```python
data_agrupada = data.groupby(['Date', 'Country/Region'])['Confirmed', 'Deaths', 'Recovered', 'actives'].sum().reset_index()
```

#### agrupou por data e região as colunas numericas de informações

```python
data_agrupada.sort_values(by='Confirmed', ascending=False)
```

```python
data_agrupada_paises = data.groupby(['Date', 'Country/Region'])['Confirmed', 'Deaths', 'Recovered', 'actives'].sum().reset_index()
```

```python
data_agrupada_tempo= data.groupby(['Date'])['Deaths', 'Recovered', 'actives'].sum().reset_index()
```

```python
data_agrupada_tempo = data_agrupada_tempo.melt(id_vars ="Date", value_vars=['Deaths', 'Recovered', 'actives'], var_name ='Case', value_name='Count')
```

```python
import plotly.express as px
```

```python
fig = px.area(data_agrupada_tempo,
             x='Date',
             y='Count',
             color='Case',
             height= 600,
             title='Casos de covid de acordo com o tempo',
             color_discrete_sequence = [recuperados, mortes, ativos]
             )
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```

```python
import numpy as np
```

```python
fig = px.choropleth(data_agrupada_paises,
                   locations ="Country/Region",
                   locationmode= 'country names',
                   color=np.log(data_agrupada_paises["Confirmed"]),
                   hover_name = 'Country/Region',
                   hover_data=["Confirmed"],
                   animation_frame=data_agrupada_paises["Date"].dt.strftime('%d-%m-%Y'),
                   title='Pesperctiva Mundial',
                   color_continuous_scale=px.colors.sequential.Magenta)
fig.update_layout(autosize=False, width=1200, height=800)                  
fig.show()
```
